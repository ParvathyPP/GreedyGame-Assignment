```python
#grouping all other features
Ad_Requests = df.groupby('App')['Ad Requests'].sum()
Ad_Responses = df.groupby('App')['Ad Responses'].sum()
Impressions = df.groupby('App')['Impressions'].sum()
First_click = df.groupby('App')['1st Clicks'].sum()
Second_click = df.groupby('App')['2nd Clicks'].sum()
```

```python
type(Second_click)
```

```python
df.columns
```

```python
#creating a new dataframe by groupby App name 
new_df = pd.concat([Revenue, Ad_Requests, Ad_Responses, Impressions, First_click, Second_click], axis = 1)
```

```python
new_df
```

```python
#Try again with the correlation to check if the revenue have any correlation with the features
plt.figure(figsize = (12, 9))
sns.heatmap(new_df.corr(), annot = True)
#Revenue has high correlation with Ad Requests, Ad responses, Impressions, 1st click and 2nd Click
```

```python
new_df.columns
```

```python
#Finding the fill rate and Render rate for the new dataframe. calculating from scratch since the fill rate data was erraneous in the first dataset
new_df['Fill_rate'] = new_df.apply(lambda row :(row['Ad Responses'] / row['Ad Requests'] * 100) if row['Ad Requests'] != 0 else 0, axis = 1)
new_df['Render_Rate'] = new_df.apply(lambda row :(row['Impressions'] / row['Ad Responses'] * 100) if row['Ad Responses'] != 0 else 0, axis = 1)
```

```python
#grouping by App name and Ad Group
Revenue_ad_group = df.groupby(['App', 'Ad Group'])['Revenue'].sum()
Revenue_ad_group.sort_values(ascending = False, inplace = True)
Ad_Requests_ad_group = df.groupby(['App', 'Ad Group'])['Ad Requests'].sum()
Ad_Responses_ad_group = df.groupby(['App', 'Ad Group'])['Ad Responses'].sum()
Impressions_ad_group = df.groupby(['App', 'Ad Group'])['Impressions'].sum()
First_click_ad_group = df.groupby(['App', 'Ad Group'])['1st Clicks'].sum()
Second_click_ad_group = df.groupby(['App', 'Ad Group'])['2nd Clicks'].sum()
```

```python
new_df_ad_group = pd.concat([Revenue_ad_group, Ad_Requests_ad_group, Ad_Responses_ad_group, Impressions_ad_group, First_click_ad_group, Second_click_ad_group], axis = 1)
```

```python
new_df_ad_group
```

```python
new_df_ad_group['Fill_rate'] = new_df_ad_group.apply(lambda row :(row['Ad Responses'] / row['Ad Requests'] * 100) if row['Ad Requests'] != 0 else 0, axis = 1)
new_df_ad_group['Render_Rate'] = new_df_ad_group.apply(lambda row :(row['Impressions'] / row['Ad Responses'] * 100) if row['Ad Responses'] != 0 else 0, axis = 1)
```
